```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('https://makeheadway.com/library/topics/financial-literacy/',headers=headers).text
```

```python
soup=BeautifulSoup(webpage)
soup.prettify()
```

```python
book=soup.find_all('div',class_='LibraryTopicsList_topic__right-column__nvGbv')
len(book)
```

```python
name=[]
auther=[]
rating=[]
discription=[]
c=0
for i in book:
    
    name.append(i.find('h2').text)
    auther.append(i.find('p').text)
    rating.append(i.find_all('span')[4].text[:3])
    discription.append(i.find_all('p')[3].text)
    c+=1
    if c==43:
        break

# name
# auther
# rating
# discription
    
```

```python
df={'name':name,'rating':rating,'auther':auther,'discription':discription}
datafbook=pd.DataFrame(df)
```
